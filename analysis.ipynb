```python
import pandas as pd
from database.adatabase import ADatabase
import matplotlib.pyplot as plt
```

```python
cash_money = ADatabase("cash_money")
sapling = ADatabase("sapling")
```

```python
cash_money.connect()
kpi = cash_money.retrieve("kpi")
cash_money.disconnect()
```

```python
strategy_kpis = kpi.sort_values("return",ascending=False).groupby(["strategy"]).first().sort_values("return",ascending=False).reset_index()
```

```python
plt.scatter(strategy_kpis["std"].values,strategy_kpis["return"])
plt.show()
```
