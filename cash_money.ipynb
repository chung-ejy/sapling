```python
from backtest_functions.backtest_functions import BacktestFunctions as bf
from strategy.strategy import Strategy
import pandas as pd
from tqdm import tqdm
import matplotlib.pyplot as plt
from database.adatabase import ADatabase
```

```python
db = ADatabase("cash_money")
names = Strategy._member_names_
queries = []
for name in tqdm(names):
    for ascending in [True,False]:
        for positions in [5,10,20]:
            for stop_loss in [0.05,1]:
                query = {
                    "strategy":name,
                    "holding_period":5,
                    "positions":positions,
                    "stop_loss":stop_loss,
                    "ascending":ascending,
                }
                queries.append(query)
```

```python
analysis = []
db.connect()
db.drop("kpi")
for query in tqdm(queries):
    try:
        results = bf.backtest(query)
        db.store("kpi",pd.DataFrame([results["kpi"] | query]))
    except Exception as e:
        print(str(e))
        continue
db.disconnect()
```
